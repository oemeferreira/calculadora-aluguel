```python
import pandas as pd
df = pd.read_csv('full_results.csv')
```

```python
df.head()
```

```python
df[df['id'].duplicated() == True]
```

```python
df = df.drop_duplicates(subset=['id'], keep='last')
```

```python
df.shape
```

```python
df.columns
```

```python
all_columns = list(df)
df[all_columns] = df[all_columns].astype(str)
```

```python
df.dtypes
```

```python
df['garages'] = df['garages'].str.extract('(.*) Vaga')
```

```python
df.head()
```

```python
df['bathrooms'] = df['bathrooms'].str.extract('(.*) Banheiro')
df['rooms'] = df['rooms'].str.extract('(.*) Quarto')
df.head()
```

```python
df['price'] = df['price'].str.extract('R\$ (.*) /mês')
df['price'] = df['price'].replace('\.', '', regex=True)
df.head()
```

```python
df['condo'] = df['condo'].str.extract('R\$ (.*)')
df['condo'] = df['condo'].replace('\.', '', regex=True)
df.head()
```

```python
df.address.str.split('\s-').str[1]
```

```python
df['bairro'] = df.address.str.split('\s-').str[1]
df['neighborhood'] = df.bairro.str.split(',').str[0]
#df['street'] = df.address.str.split('\s-').str[0]
df = df.drop(['bairro'],axis=1)
df.head()
```

```python
df['neighborhood'].unique()
```

```python
df.to_csv('full_limpo.csv')
```
